```python
con = sqlite3.connect('w5.db')
cur = con.cursor()
```

```python
def select():
    cur.execute('SELECT * FROM subjects')
    rows = cur.fetchall()
    return rows
```

```python
def avail_subjects():
    rows = select()
    avail = []
    for i in rows:
        avail.append((i[1].strip(),i[2].strip(),i[3],i[7]))
    return set(avail)
```

```python
avail_sub = list(avail_subjects())
avail_sub
```

```python
def remove_dupe(ava):
    avail_sub = ava
    to_remove = []
    replace = []
    for i in avail_sub:
        if len(i[0]) < 9:
            to_remove.append(i)
            replace.append([
                '0'+i[0],
                i[1],
                i[2],
                i[3]
            ])
    for i in to_remove:
        avail_sub.remove(i)
    for i in replace:
        avail_sub.append(tuple(i))
    avail_sub = set(avail_sub)
    sub_dict = {}
    for i in avail_sub:
        if i[0] not in sub_dict:
            sub_dict[i[0]] = (i[1],i[2],i[3])
        else:
            if len(i[1]) < len(sub_dict[i[0]]):
                sub_dict[i[0]] = (i[1],i[2],i[3])
    return sub_dict
```

```python
remove_dupe(avail_sub)
```
